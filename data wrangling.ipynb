```python
import numpy as np
import pandas as pd
```

```python
dependent_var = ['RUL']
index_columns_names =  ["id","Cycle"]
operational_settings_columns_names = ["setting"+str(i) for i in range(1,4)]
sensor_measure_columns_names =["s"+str(i) for i in range(1,22)]
input_file_column_names = index_columns_names + operational_settings_columns_names + sensor_measure_columns_names
```

```python
df_train = pd.read_csv('data/data_train.txt',delim_whitespace=True,names=input_file_column_names)

rul = pd.DataFrame(df_train.groupby('id')['Cycle'].max()).reset_index()
rul.columns = ['id', 'max']
df_train = df_train.merge(rul, on=['id'], how='left')
df_train['ttf'] = df_train['max'] - df_train['Cycle']
df_train.drop('max', axis=1, inplace=True)

df_train.head()
```

```python

def prepare_test_data(df_test_in, df_truth_in, period):
    
    df_tst_last_cycle = pd.DataFrame(df_test_in.groupby('id')['Cycle'].max())
    
    df_tst_last_cycle.reset_index(level=0, inplace=True)
    df_tst_last_cycle.columns = ['id', 'last_cycle']
    
    df_test_in = pd.merge(df_test_in, df_tst_last_cycle, on='id')


    df_test_in = df_test_in[df_test_in['Cycle'] == df_test_in['last_cycle']]

    df_test_in.drop(['last_cycle'], axis=1, inplace=True)
    
    df_test_in.reset_index(drop=True, inplace=True)
    
    df_test_in = pd.concat([df_test_in, df_truth_in], axis=1)
    return df_test_in
```

```python
data_test = prepare_test_data(df_test, RUL, 15)
data_test.head()
```
